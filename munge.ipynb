```python
import json
import unicodedata
import random
import spacy

from collections import Counter, defaultdict
from dotenv import dotenv_values

nlp = spacy.load('en_core_web_sm')
config = dotenv_values(".env")
```

```python
with open('./jokes.json') as f:
    data = json.load(f)
```

```python
jokes = [d['body'] for d in data]
```

```python
jokes_by_len = defaultdict(list)
counts = defaultdict(int)

for joke in jokes:
    if 'Conan' in joke:
        continue
    
    joke_clean = unicodedata.normalize("NFKD", joke).replace('  ', ' ').strip()

    tokens = nlp(joke_clean)
    sentences = [sent.text for sent in tokens.sents]
    sentence_ct = len(sentences)
    
    jokes_by_len[sentence_ct].append(joke_clean)
    counts[sentence_ct] += 1

print(counts)
```

```python
# check sample of jokes with given num of sentences
sentence_ct = 1
sample_n = 10

print('\n\n'.join(random.sample(jokes_by_len[sentence_ct], sample_n)))
```
